```python
import pandas as pd
import numpy as np
 
N=20

df = pd.DataFrame({
    'A': pd.date_range(start='2016-01-01',periods=N,freq='D'),
    'x': np.linspace(0,stop=N-1,num=N),
    'y': np.random.rand(N),
    'C': np.random.choice(['Sci','Com','Hum'],N).tolist(),
    'D': np.random.normal(100, 10, size=(N)).tolist()
    })

#for col in df:
#   print(col)
print(df)
```

```python
for i in df.iteritems():
    print(i)
```

```python
for i,j in df.iteritems():
    print(i,j)
```

```python
for i,j in df.iterrows():
    print(i)
```

```python
for i,j in df.iterrows():
    print(j)
```

```python
for i in df.itertuples():
    print(i)
```
